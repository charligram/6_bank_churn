# Feature engineering

## Inicio

#### Librerías

```python
import pandas as pd
```

#### Importar datos

```python
DATA_ROUTE = r"../data/raw/Churn_Modelling.csv"
```

```python
df = pd.read_csv(DATA_ROUTE)
```

## Trabajo con features

#### Crear AgeGroup

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[
        18,
        30,
        45,
        60,
        93
    ],
    labels=[
        "Young adult",
        "Adult",
        "Middle Age",
        "Senior"
    ],
    right="False"
)
```

#### Crear HasBalance

```python
df["HasBalance"] = (df["Balance"] > 0).astype(int)
```

## Guardar csv

```python
df.to_csv("../data/processed/Churn_Modelling_Processed.csv", index=False)
```
