## Inicio

#### Librerías e imports

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from src.feature_engineering import generate_age_numproducts_score, generate_balance_and_active_score, generate_AgeAndSalary_score
```

#### Importar datos

```python
DATA_ROUTE = r"../data/raw/Churn_Modelling.csv"
```

```python
df = pd.read_csv(DATA_ROUTE)
```

## Trabajo con features

#### Crear AgeGroup

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[
        18,
        30,
        45,
        60,
        93
    ],
    labels=[
        "Young adult",
        "Adult",
        "Middle Age",
        "Senior"
    ],
    right=False
)
```

#### Crear HasBalance

```python
df["HasBalance"] = (df["Balance"] > 0).astype(int)
```

#### Crear Age_x_NumOfProducts_score

```python
df["Age_x_NumOfProducst_score"] = df.apply(
    generate_age_numproducts_score,
    axis=1 # Actuar en cada fila
)
```

#### Crear BalanceAndActive

```python
df["BalanceAndActive"] = df.apply(
    generate_balance_and_active_score,
    axis=1
)
```

#### Crear Score_AgeAndSalaryScore

```python
df["Score_AgeAndSalaryScore"] = df.apply(
    generate_AgeAndSalary_score,
    axis=1
)
```

## Guardar csv

```python
df.to_csv("../data/processed/Churn_Modelling_Processed.csv", index=False)
```
